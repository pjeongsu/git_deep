```python
# 데이터를 읽어온다
df1 = pd.read_csv('../../data/naver_star_data.csv')
df1
```

```python
# 평점과 평가글만 가져온다
df2 = df1[['평점', '평가글']]
df2.head()
```

```python
# 결측치는 문자열이 아닌 실수타입으로 인식해서
# 아예 결측치를 제거하자

df2.dropna(inplace=True)
df2.isna().sum()
```

```python
# 평점이 5점 이하는 0으로 바꿔준다
a1 = df2.query('평점 <= 5').index
df2.loc[a1, '평점'] = 0
```

```python
# 평점이 6점 이상은 1로 바꿔준다.
a1 = df2.query('평점 > 5').index
df2.loc[a1, '평점'] = 1
```

```python
df2['평점'].value_counts()
```

```python
# 평가글 추출
docs = df2['평가글'].values
docs
```

```python
# 결과 추출
classes = df2['평점'].values
classes
```

```python
# 토큰화
token = Tokenizer()
token.fit_on_texts(docs)
token.word_index
```

```python
len(token.word_index)
```

```python
# 단어사전을 통해 문장의 각 단어를 숫자로 변환한다
x = token.texts_to_sequences(docs)
len(x)
```

```python
# 한 문장의 최대 단어수를 가져온다
max_count = 0

for x2 in x:
    # 현재 문장의 글자수
    count1 = len(x2)
    # 현재 문장의 글자수가 이전 최대 수치보다 많으면 덮어씌운다
    if max_count < count1:
        max_count = count1
        
max_count
```

```python
# 패딩 : 서로 길이가 다른 리스트의 개수를 맞춰준다
padded_x = pad_sequences(x, max_count)
padded_x
```

```python
# 단어의 개수를 구한다
word_size = len(token.word_index) + 1
word_size
```

```python
model = Sequential()
model.add(Embedding(word_size, 8, input_length=48))

model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(padded_x, classes, epochs=20)
```

```python
# 학습 데이터에 대한 성능평가

model.evaluate(padded_x, classes)[1]
```
