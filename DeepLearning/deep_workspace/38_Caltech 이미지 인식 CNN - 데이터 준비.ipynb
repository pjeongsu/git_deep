### 이미지 크기 동일하게 색상도 RGB로 동일하게!

- 색상이 있는 이미지를 검색한다.
- 본 파일에서는 딥러닝을 하기 전에 이미지 전처리 작업을 한다.
- 이미지의 크기가 모두 달라서 머신러닝으로 다루기가 어렵다.
- 일정한 크기로 새 이미지를 만들고 24비트 RGB 컬러 형태로 변환하여 모든 이미지의 크기와 색감을 통일시킨다.
- 그리고 머신러닝을 쉽게 하기 위해 ndarray형태로 변환한 다음, 이것을 파일로 저장한다.

```python
from PIL import Image
import os
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# 하위에 있는 폴더들의 목록을 가져온다.
categories = list(os.walk('img/101_ObjectCategories'))[0][1]
categories
```

```python
# 이미지 파일들을 가지고 있는 경로의 루트
caltech_dir = 'img/101_ObjectCategories'
# 전체 카테고리의 개수
nb_class = len(categories)
```

```python
# 이미지의 크기
image_w = 64
image_h = 64
pixels = image_w * image_h * 3 # RGB
```

```python
import glob

# 이미지 변형
X = []
Y = []
# 반복문
for idx, cat in enumerate(categories) :
    # print(idx, cat)
    # 이미지의 개수에 만큼의 0을 채운 리스트를 생성한다..
    label = [0 for i in range(nb_class)]
    # print(label)
    # 현재 번째의 값을 1로 채워준다.
    label[idx] = 1
    # 읽어올 이미지 경로
    image_dir = caltech_dir + '/' + cat
    # 이미지 파일
    # 해당 경로에 있는 모든 이미지 파일들의 목록을 가져온다.
    files = glob.glob(image_dir + '/*.jpg')
    # print(files)
    # 파일의 수만큼 반복한다.
    for i, f in enumerate(files) :
        # 이미지를 읽어온다.
        img = Image.open(f)
        # 24bit RGB로 변환
        img = img.convert('RGB')
        # 이미지 크기 조정
        img = img.resize((image_w, image_h))
        # 행렬로 변환
        data = np.asarray(img)
        # 리스트에 담는다.
        X.append(data)
        Y.append(label)
        # 10회마다 한번씩 출력한다.
        if i % 10 == 0 :
            print(i, '\n', data)

X = np.array(X)
Y = np.array(Y)
```

```python
# 학습용과 검증용으로 나눈다.
X_train, X_test, y_train, y_test = train_test_split(X, Y)
# 하나로 합친다.
xy = (X_train, X_test, y_train, y_test)
# 저장한다.
np.save('img/caltech_obj_101.npy', xy)
```
