```python
# 사용할 이미지가 들어 있는 폴더 위치
root_dir = 'img/101_new'
```

```python
# 폴더 이름들을 가져와 결과데이터로 구성한다.
a1 = os.walk(root_dir)
categories = list(a1)[0][1]
categories
```

```python
# 결과 데이터의 개수를 담는다.
nb_classes = len(categories)
nb_classes
```

- 입력 데이터 : 사진
- 결과 데이터 : 카테고리 이름

```python
# 입력과 결과데이터를 담을 리스트
# 입력 데이터 (이미지 데이터)
X = []
# 결과 데이터 (폴더이름을 숫자로 변환한것)
Y = []
```

```python
# 제공되는 이미지의 크기가 굉장히 다양하므로 한가지 사이즈로
# 변환을 해야 한다.
# 이 때, 변환될 가로 세로 사이즈
image_w = 64
image_h = 64
```

```python
# 폴더의 수만큼 반복한다.
for idx, cat in enumerate(categories) :
    # 이미지를 읽어올 폴더의 이름을 구성한다.
    image_dir = root_dir + '/' + cat
    # print(image_dir)
    # 해당 폴더에 있는 파일 이름들을 가져온다.
    files = glob.glob(image_dir + '/*')
    # print(files)
    
    print(f'{cat} 처리중')
    
    # 파일의 수 만큼 반복한다.
    for i, f in enumerate(files) :
        # 파일의 데이터를 읽어온다.
        img = Image.open(f)
        # 색상 모드를 변경한다.
        img = img.convert('RGB')
        # 이미지의 크기를 변경한다.
        img = img.resize((image_w, image_h))
        # plt.imshow(img)
        # plt.show()
        # 이미지 데이터를 ndarray 형태로 생성한다.
        data = np.asarray(img)
        # print(data)
        
        # 데이터를 담는다
        X.append(data)
        Y.append(idx)
```

```python
# ndarray를 생성한다.
X = np.array(X)
Y = np.array(Y)
```

```python
# 학습 데이터와 테스트용 데이터로 나눈다.
X_train, X_test, y_train, y_test = train_test_split(X, Y)
```

```python
# 저장한다.
xy = (X_train, X_test, y_train, y_test)
np.save('img/calthec_new.npy', xy)
print(f'저장완료 : {len(Y)}')
```
