```python
# seed 설정
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 입력
df1 = pd.read_csv('../../data/wine.csv', header=None)
df1.head()
```

```python
# 데이터프레임 일부를 발췌한다
# df1 = df1.sample(n=1000).reset_index(drop=True)
# print(df1)

# 데이터 분류
x = df1.drop(12, axis=1)
y = df1[12]

display(x)
display(y)
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(30, input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=200, batch_size=5)
```

```python
# 정확도
# 학습용 데이터이기 때문에 과적합이 발생했는지 안했는지 모름

model.evaluate(x, y)[1]
```

### 학습된 거 중에 epoch가 193일 때의 모델을 가져오고 싶다
#### epoch는 순전파로 갔다가 다시 역전파로 갔을 때 1 올라감

```python
# 모델 저장
# 마지막 epoch의 가중치와 바이어스가 들어간 모델이 저장됨

model.save('test1.h5')
```

```python
# 모델 불러오기(로딩)

model2 = load_model('test1.h5')
model2

# 객체가 잘 복원되었음을 알 수 있음
```
