- 영상이나 이미지에서 코너를 검출하는 알고리즘
- 정확하게 트래킹하기 좋은 지점을 찾을 수 있다.
- 다각형 등에서 꼭지점을 찾는 목적으로 사용한다.

```python
import cv2
```

```python
src = cv2.imread('opencv_image/19.jpg', cv2.IMREAD_REDUCED_COLOR_2)
```

```python
# 작업을 위해 이미지를 복제한다.
dst = src.copy()
```

```python
# 그레이 스케일
gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
```

```python
# 코너를 검출한다.
# 그레이스케일이 적용된 이미지, 코너 최댓값, 코너 품질, 코너간 최소 거리
# 마스크, 블록크기
corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 5)
```

```python
corners
```

```python
for i in corners :
    # 원으로 표시한다
    # 원의 중심점 좌표, 선의 두께, 색상, 반경
    cv2.circle(dst, tuple(i[0]), 0, (0, 0, 255), 2)
```

```python
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
