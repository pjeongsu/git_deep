# PCA
### 변동량이 가장 많은 칼럼을 중심으로 압축

```python
# n_components : 칼럼수 지정

pca = PCA(n_components=2)

x = iris_df.drop(4, axis=1)

# 학습
pca.fit(x)

# 데이터를 변환
r1 = pca.transform(x)
r1
```

```python
pca_df = pd.DataFrame(r1, columns=['p1', 'p2'])
pca_df['target'] = iris_df[4]
pca_df.head()
```

```python
sns.pairplot(pca_df, hue='target')
plt.show()
```

- 시각화를 통해 결과를 잘 예측할 수 있겠구나 확인할 수 있음
