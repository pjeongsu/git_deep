- h12_g3 : 성별(1 : 남자, 2 : 여자, 9 : 모름/무응답)
- h12_g4 : 태어난 년도(1900 ~ 2014, 9999 : 모름/무응답)
- h12_g10 : 혼인상태(0 : 미해당, 1 : 유배우, 2 : 사별, 3 : 이혼, 4 : 별거, 5 : 미혼(18세 이상, 미혼모 포함), 6 : 기타, 9 : 모름/무응답)
- h12_g11 : 종교(1 : 있음, 2 : 없음, 9 : 모름/무응답)
- h12_eco9 : 직종(직종코드표, 9999 : 모름/무응답)
- p1202_8aq1 : 일을 한 달의 월 평균 임금(1 ~ 9998, 만원 단위, 9999 : 모름/무응답)
- h12_reg7 : 7개 권역별 지역 부분(1 : 서울, 2 : 수도권(인천/경기), 3 : 부산/경남/울산, 4 : 대구/경북, 5 : 대전/충남, 6 : 강원/충북, 7 : 광주/전남/전북/제주도)


- 조사 대상에서 남녀 성별 비를 파악한다
- 성별에 따른 평균 월급 차이 분석
- 나이에 따른 평균 임금의 변화
- 종교 유무에 따른 이혼 비율

```python
# 경고 메세지가 안나오게..
import warnings
warnings.filterwarnings('ignore')

# 기본
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 그래프 설정
plt.rcParams['font.family']='Malgun Gothic'
# 맥용 plt.rcParams['font.family']='AppleGothic'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['axes.unicode_minus'] = False
```

```python
df1 = pd.read_excel('../data/dataset2017.xlsx', )
df1.head()
```

```python
df = df1[['h12_g3', 'h12_g4', 'h12_g10', 'h12_g11', 'h12_eco9', 'p1202_8aq1', 'h12_reg7']]
df.rename({'h12_g3':"성별", 'h12_g4':"출생년도", 'h12_g10':"혼인상태", 'h12_g11':"종교", 'h12_eco9':"직종",'p1202_8aq1':"월평균임금", 'h12_reg7':"지역"}, axis = 1, inplace = True)
df
```

```python
df.isna().sum()
```

```python
df['월평균임금'] = df['월평균임금'].fillna(9999)
df['직종'] = df['직종'].fillna(9999)
```

```python
df.isna().sum()
```

# 1. 조사 대상에서 남녀 성별 비

```python
df['성별'].unique()
```

```python
df_sex = pd.DataFrame(df['성별'].value_counts())
df_sex.sort_index(inplace=True)
df_sex
```

```python
df_sex['비율(%)'] = round((df_sex['성별']/df['성별'].shape[0])*100, 2)
df_sex.rename({'성별':"명수"}, axis = 1, inplace = True)
df_sex.index = ['남자', '여자']
df_sex
```

```python
explode = [0.04, 0.04]
plt.pie(df_sex['명수'], labels=df_sex.index, autopct='%.2f%%', explode=explode, shadow=True)
plt.title('남녀 성비')
plt.show()
```

# 2. 성별에 따른 평균 월급 차이 분석


```python
df_sal = df.copy()
```

```python
inx_del = df_sal[df_sal['월평균임금'] == 9999].index
df_sal = df_sal.drop(inx_del)
```

```python
df_sal.groupby('성별')[['월평균임금']].mean()
```

```python
df_sal.groupby('성별')[['월평균임금']].max()
```

```python
df_sal.groupby('성별')[['월평균임금']].min()
```

```python
man_sal = df_sal[df_sal['성별'] == 1][['월평균임금']]
woman_sal = df_sal[df_sal['성별'] == 2][['월평균임금']]
display(man_sal)
display(woman_sal)
```

```python
sal = [man_sal, woman_sal]
```

```python
plt.boxplot(sal)
plt.title('성별에 따른 평균 월급')
plt.show()
```

# 3. 나이에 따른 평균 임금의 변화: 출생년도를 계산(2017년 기준으로 하기)


```python
df_sal['나이(2017년 기준)'] = 2017 - df_sal['출생년도'] + 1
df_sal
```

```python
plt.scatter(x='나이(2017년 기준)', y='월평균임금', data=df_sal)
plt.title('나이별 평균 임금 산점도', fontsize=25)
plt.xlabel('나이')
plt.ylabel('평균 임금(만원)')
plt.show()
```
