# 데이터를 읽어온다

```python
df1 = pd.read_csv('../data/iris.csv', header=None)
df1.head()
```

# 입력과 결과로 나눈다

```python
x = df1.drop(4, axis=1)
y = df1[4]

display(x)
display(y)
```

# 군집 수행

```python
# n_clusters : 군집의 개수, 센트로이드의 개수
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
```

```python
# 군집 결과를 가져온다
kmeans.labels_
```

```python
# 진짜 결과와 예측 결과를 가진 데이터프레임을 각각 생성한다
original_df = x.copy()
predition_df = x.copy()

original_df['target'] = y
predition_df['target'] = kmeans.labels_

display(original_df.head())
display(predition_df.head())
```

- pairplot으로 패턴 확인하기

```python
sns.pairplot(original_df, hue='target')
plt.show()
```

```python
sns.pairplot(predition_df, hue='target')
plt.show()
```

```python
# 결과 비교를 위해 진짜 결과를 숫자로 변환
map1 = {
    'Iris-setosa' : 0,
    'Iris-versicolor' : 1,
    'Iris-virginica' : 2
}

original_df['target'] = original_df['target'].map(map1)
original_df['target']
```

```python
r1 = accuracy_score(original_df['target'], predition_df['target'])
r1
```
