```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data
import datetime
import seaborn as sns
```

```python
# 그래프 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 시작날짜
start = datetime.datetime(2010, 1, 1)
# 종료날짜
end = datetime.datetime(2020, 12, 31)
```

```python
# 삼성전자 주식 데이터를 가져온다
f = data.DataReader('005930.KS', 'yahoo', start, end)
f
```

```python
# seaborn에서 정의한 그래프 스타일을 적용한다
sns.set()
# 시각화
f.plot()
plt.show()
```

```python
f['High'].plot()
plt.show()
```

```python
# S : 매초, H : 매시간, A : 매년, T : 분단위, D : 일단위, W : 주단위, M : 월단위, Y : 년단위
list(f.resample('A'))[0]
```

```python
list(f.resample('Y'))
```

```python
f['High'].resample('Y').mean().plot()
plt.show()
```

```python
f.asfreq('Y', method='ffill')
```

```python
# 19040.0	18840.0	18840.0	18980.0	9181500.0	15656.641602
a1 = f.asfreq('Y', method='ffill')
a2 = a1[['Close', 'High', 'Low']]
a2.plot()
plt.show()
```
